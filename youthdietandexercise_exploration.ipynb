```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st


import json
import requests
from pprint import pprint
```

```python
file_path = "Resources/Chicago_Health_Youth_activity.csv"
youth_activity_df= pd.read_csv(file_path)
youth_activity_df
```

```python
cleaned_youth_activity_count = youth_activity_df.drop(index=[0,2],
                                                      columns=['Layer','Name','GEOID','YRPAP_2019','YRPAP_2017',
                                                               'YRPAP_2013','YRPAP_2011','YRPAP_2009','YRPAP_2007',
                                                              'YRPAP_2005','YRPA-M_2019'])

cleaned_youth_activity_count=cleaned_youth_activity_count[cleaned_youth_activity_count.columns[::-1]]
cleaned_youth_activity_count=cleaned_youth_activity_count.astype(float)
final_youth_activity_count=cleaned_youth_activity_count.rename(columns={'YRPA_2005':'2005',
                                                                 'YRPA_2007':'2007','YRPA_2009':'2009','YRPA_2011':'2011',
                                                                 'YRPA_2013':'2013','YRPA_2017':'2017','YRPA_2019':'2019'})
final_youth_activity_count
```

```python
file_path2 = "Resources/Chicago_Health_Atlas_youth_diet.csv"
youth_diet_df= pd.read_csv(file_path2)
youth_diet_df
```

```python
cleaned_youth_diet_count = youth_diet_df.drop(index=[0,2],
                                                      columns=['Layer','Name','GEOID','YRFVP_2019','YRFVP_2017',
                                                               'YRFVP_2013','YRFVP_2011','YRFVP_2009','YRFVP_2007',
                                                              'YRFVP_2005','YRFVP_2003','YRFVP_2001'])

cleaned_youth_diet_count=cleaned_youth_diet_count[cleaned_youth_diet_count.columns[::-1]]
cleaned_youth_diet_count=cleaned_youth_diet_count.astype(float)
final_youth_diet_count=cleaned_youth_diet_count.rename(columns={'YRFV_2001':'2001','YRFV_2003':'2003','YRFV_2005':'2005',
                                                                 'YRFV_2007':'2007','YRFV_2009':'2009','YRFV_2011':'2011',
                                                                 'YRFV_2013':'2013','YRFV_2017':'2017','YRFV_2019':'2019'})
final_youth_diet_count
```

```python

file_path3 = "Resources/Chicago_Health_Atlas_youth_drinking.csv"
youth_drinking_df= pd.read_csv(file_path3)
youth_drinking_df
```

```python
cleaned_youth_drinking_count = youth_drinking_df.drop(index=[0,2],
                                                      columns=['Layer','Name','GEOID','YRBDP_2019','YRBDP_2017'])

cleaned_youth_drinking_count=cleaned_youth_drinking_count[cleaned_youth_drinking_count.columns[::-1]]
cleaned_youth_drinking_count=cleaned_youth_drinking_count.astype(float)
final_youth_drinking_count=cleaned_youth_drinking_count.rename(columns={'YRBD_2017':'2017','YRBD_2019':'2019'})
final_youth_drinking_count
```

```python
#Put everything into one dataframe to save to csv
frames = [final_youth_diet_count,final_youth_activity_count,final_youth_drinking_count]

youth_diet_exercise_alcohol_df = pd.concat(frames)
youth_diet_exercise_alcohol_df=youth_diet_exercise_alcohol_df.reset_index(drop=True)
youth_diet_exercise_alcohol_df=youth_diet_exercise_alcohol_df.rename(index={0:'YRFV',1:'YRPA',2:'YRBD'})
youth_diet_exercise_alcohol_df
```

```python
#Save to csv
filepath = 'Resources/youth_diet_exercise_alcohol.csv'
youth_diet_exercise_alcohol_df.to_csv(filepath)
```
